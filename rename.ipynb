```python
import os
import csv

# Paths to the directories and CSV file
manual_path = "E:\\A files\\Master\\Dissertation\\Samples\\test1"
t2t_path = "E:\\A files\\Master\\Dissertation\\Samples\\test2"
csv_file = "E:\\A files\\Master\\Dissertation\\Samples\\MLA01_name_map.csv"
output_csv = "E:\\A files\\Master\\Dissertation\\Samples\\renamed_files.csv"

renamed_files = []  # 用于存储旧文件名和新文件名的列表

with open(csv_file, 'r') as f:
    reader = csv.reader(f)
    
    # Skip the header
    next(reader)
    
    rename_counter = 0
    for row in reader:
        # Prepare file names
        old_name_manual = os.path.join(manual_path, row[0] + ".ply")
        new_name_manual = os.path.join(manual_path, f'T{rename_counter+1}_M.ply')
        old_name_t2t = os.path.join(t2t_path, row[1] + ".leafon.ply")
        new_name_t2t = os.path.join(t2t_path, f'T{rename_counter+1}_T.ply')
        
        # Check if both files exist before renaming
        if os.path.isfile(old_name_manual) and os.path.isfile(old_name_t2t):
            try:
                # Rename files in the manual directory
                os.rename(old_name_manual, new_name_manual)
                print(f'Renamed {old_name_manual} to {new_name_manual}')
                renamed_files.append((os.path.splitext(os.path.basename(old_name_manual))[0], os.path.splitext(os.path.basename(new_name_manual))[0]))
                
                # Rename files in the t2t directory
                os.rename(old_name_t2t, new_name_t2t)
                print(f'Renamed {old_name_t2t} to {new_name_t2t}')
                renamed_files.append((os.path.splitext(os.path.basename(old_name_t2t))[0], os.path.splitext(os.path.basename(new_name_t2t))[0]))
                
                rename_counter += 1  # Only increment counter if renaming was successful
            except Exception as e:
                print(f'Error occurred while renaming files: {str(e)}')
        else:
            print(f'Skipping as one or both files are missing for group {row[0]} and {row[1]}.')

# 将重命名的文件列表写入新的CSV文件
with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Old Name', 'New Name'])  # 写入标题行
    for old_name, new_name in renamed_files:
        writer.writerow([old_name, new_name])
```
