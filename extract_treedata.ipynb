```python
import pandas as pd
from scipy.io import loadmat
import os

def extract_treedata_from_mat(mat_file_path):
    try:
        mat_data = loadmat(mat_file_path)
        qsm_data = mat_data.get('qsm', {})[0][0]
        treedata = qsm_data['treedata'][0][0]
        
        parameters = {}
        for key in treedata.dtype.fields.keys():
            parameters[key] = treedata[key][0][0]
            
        return parameters
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def main():
    # Read name_list.csv to get the list of .mat files to process
    name_list_df = pd.read_csv("E:\\A files\\Master\\Dissertation\\Samples\\WYT\\name_list.csv", header=None)
    mat_file_names = name_list_df[0].tolist()
    
    # Initialize an empty DataFrame to store the extracted parameters
    df = pd.DataFrame()
    
    # Loop through each .mat file and extract the parameters
    for mat_file_name in mat_file_names:
        mat_file_path = f"E:\\A files\\Master\\Dissertation\\Samples\\WYT\\ww_leafoff_models\\{mat_file_name}.mat"
        
        if os.path.exists(mat_file_path):
            parameters = extract_treedata_from_mat(mat_file_path)
            if parameters is not None:
                parameters['File_Name'] = mat_file_name
                df = df.append(parameters, ignore_index=True)
        else:
            print(f"File {mat_file_name}.mat not found.")
    
    # Save the extracted parameters to a CSV file
    df.to_csv("E:\\A files\\Master\\Dissertation\\Samples\\WYT\\extracted_treedata.csv", index=False)

if __name__ == "__main__":
    main()
```
